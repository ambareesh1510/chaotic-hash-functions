```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
def analyze_hash(name):
    with open('../benchmark_data_' + name + '.csv', newline='') as csvfile:
        data = list(csv.reader(csvfile))
    print("analyzing ", name)
    for i in range(len(data[0])):
        data[0][i] = int(data[0][i])
    print(len(data[0]))
    data[1][0] = int(data[1][0])
    print("Collisions: ", data[1][0])
    print("Collision percentage: ", data[1][0] / len(data[0]) * 100, '%')
    for i in range(len(data[2])):
        data[2][i] = int(data[2][i])
        
    # plt.figure()
    plt.title(name + ': hash(n)')
    plt.hist(data[0], bins=100);
    
    plt.figure()
    plt.title(name + ': deviation(n)')
    plt.hist(data[2], bins=100, color='green');
```

```python
analyze_hash("rolling_hash")
```

```python
analyze_hash("tent_map_hash")
```

```python
analyze_hash("dyadic_map_hash")
```

```python
analyze_hash("multi_state_hash")
```
